Looks like there are likely duplicate rows in the vcbb, leaving dupes for now

```python
def color(df_):

    styles_df = pd.DataFrame('', index=df_.index, columns=df_.columns)

    styles_df.loc[df_['Service_Speed'] != df_['old_Service_Speed'], 'Service_Speed'] = 'background-color: yellow'

    styles_df.loc[df_['Status'] != df_['old_Status'], 'Status'] = 'background-color: yellow'

    return styles_df
```

```python
cols_to_keep = phase_1_df.columns.tolist()

col_update = {
    'bb_status': 'Service_Speed',
    'vcbb_status': 'Status',
    'sitetype': 'Site_Type',
    'Service_Speed': 'old_Service_Speed',
    'Status': 'old_Status',
}

temp_p1_merged = temp_p1_merged.rename(columns=col_update)

cols_to_keep.extend(col_update.values())

temp_p1_merged = temp_p1_merged[cols_to_keep]
```

```python
# write to temp file so that we can highlight cols some bug pops up if we try via chain
temp_p1_merged.to_csv('temp.csv', index=False)

df = pd.read_csv('temp.csv', dtype={'Zip Code': str})

with pd.ExcelWriter(DATA_DIR / 'processed' / 'phase_1_vcbb_update.xlsx') as writer:
    df.style.apply(color, axis=None).to_excel(writer, index=False)
```

```python
cols_to_keep = pending_phase_df.columns.tolist()

col_update = {
    'bb_status': 'Service_Speed',
    'vcbb_status': 'Status',
    'sitetype': 'Site_Type',
}

temp_pending_merged = temp_pending_merged.rename(columns=col_update)

cols_to_keep.extend(col_update.values())

temp_pending_merged = temp_pending_merged[cols_to_keep]

temp_pending_merged.to_excel(DATA_DIR / 'processed' / 'pending_phase_vcbb_update.xlsx', index=False)
```
